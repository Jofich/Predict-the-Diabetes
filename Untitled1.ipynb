```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("train.csv")
```

```python
test = pd.read_csv("test.csv")
id_test = test['Id']
```

```python
train.head()
```

```python
train = train.drop(columns = "Id")
test = test.drop(columns = "Id")
```

```python
X = train.drop(columns = "Outcome")
y = train['Outcome']
```

```python
train['Outcome'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE

os = SMOTE(random_state = 123, k_neighbors = 5)

column = X.columns

X_os, y_os = os.fit_resample(X, y)
X = pd.DataFrame(data = X_os, columns = column)
y = pd.DataFrame(data = y_os, columns = ['Outcome'])

y.value_counts()
```

```python
X
```

```python
from sklearn.svm import SVC

svm_rbf = SVC(kernel='rbf', random_state=1, gamma=0.10, C=10.0)
svm_rbf.fit(X, y)
```

```python
y_pred_svm = svm_rbf.predict(test)
```

```python
y_pred_svm
```
